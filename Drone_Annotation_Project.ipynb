```python
import cv2
import matplotlib.pyplot as plt
import random

# Pick a random image from the dataset
image_info = random.choice(coco_data["images"])
image_path = f"/content/images/{image_info['file_name']}"  # Ensure path matches

# Load the image
image = cv2.imread(image_path)
if image is None:
    print(f"Error: Image not found at {image_path}")
else:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Draw bounding boxes
    for ann in coco_data["annotations"]:
        if ann["image_id"] == image_info["id"]:
            x, y, w, h = map(int, ann["bbox"])
            cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Show the image with bounding boxes
    plt.imshow(image)
    plt.axis("off")
    plt.title("COCO Annotations")
    plt.show()
```

```python
import random
import shutil
import os

# Create train/test folders
os.makedirs("/content/images/train", exist_ok=True)
os.makedirs("/content/images/test", exist_ok=True)

# Get image filenames from COCO JSON
image_files = [img["file_name"] for img in coco_data["images"]]
random.shuffle(image_files)

# Split dataset (80% Train, 20% Test)
split_index = int(0.8 * len(image_files))
train_files, test_files = image_files[:split_index], image_files[split_index:]

# Move images to train/test folders
for file in train_files:
    shutil.move(f"/content/images/{file}", "/content/images/train/")

for file in test_files:
    shutil.move(f"/content/images/{file}", "/content/images/test/")

print(f"‚úÖ Split dataset: {len(train_files)} train images, {len(test_files)} test images.")
```
